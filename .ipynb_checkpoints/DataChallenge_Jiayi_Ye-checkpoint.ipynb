## Job 2: process and analyze the data
### Part 1 Monthly Occupancy
Now we solve the monthly unit occupency problem:
Join table on Unit Number because unit number is unique.
clean data drop row if there is an empty element in that row.

```python
RM_UM_on_UnitNumber = pd.merge(RM,UM,how='left',on='UnitNumber').dropna()
RM_UM_on_UnitNumber['StartDate'] = pd.to_datetime(RM_UM_on_UnitNumber['StartDate'])
RM_UM_on_UnitNumber['EndDate'] = pd.to_datetime(RM_UM_on_UnitNumber['EndDate'])

print (RM_UM_on_UnitNumber.shape)
```

Then we can start to process the data and format it as we want 

```python
earliest = datetime.strptime('03/01/2009', '%m/%d/%Y').date()
latest = datetime.strptime('03/01/2017', '%m/%d/%Y').date()
# Generate a month list to compare the value inside the table.
month = pd.date_range(earliest, latest, freq='MS').strftime("%m/%Y").tolist()

unit_type=['STD','1BR','2BR']
```

```python
Q1_list = []
for m in month:
#     m = m.date()
    m = datetime.strptime(m, '%m/%Y').date()
    m_list = [0,0,0]
    for index, row in RM_UM_on_UnitNumber.iterrows():
        start = (row['StartDate']).date()
        end = (row['EndDate']).date()
        if row['UnitType']=='STD' and m>=start and m<=end :
            m_list[0]+=1
        if row['UnitType']=='1BR' and m>=start and m<=end :
            m_list[1]+=1
        if row['UnitType']=='2BR' and m>=start and m<=end :
            m_list[2]+=1
    Q1_list.append(m_list)
headers = ['STD','1BR','2BR']
Q1 = pd.DataFrame(Q1_list,columns=headers)
Q1['Month'] = month
```

### Part 2 Move in, out, renewal

```python
# Q2: MOVE IN / MOVE OUT AT 
Q2 = {}
for item in month:
    Q2[item] = [0,0,0]
    
# remove duplicates to only consider unique data in the dataframe
dup_list=pd.concat(g for _, g in RM.groupby("CustomerNo") if len(g) > 1)
unique_list=RM.drop_duplicates(subset='CustomerNo',keep='first', inplace=False)
```

Move in and move out

```python
#Move in and move out
for index,row in unique_list.iterrows():
    moveIn = datetime.strptime(row['StartDate'], '%m/%d/%Y').date()
    moveIn = moveIn.strftime("%m/%Y")
    if type(row['EndDate']) == float:
        moveOut = '02/2019'
    else:
        moveOut = datetime.strptime(row['EndDate'], '%m/%d/%Y').date()
        moveOut = moveOut.strftime("%m/%Y")

    if moveIn in Q2:
        Q2[moveIn][0] += 1
    else:
        Q2[moveIn] = [1,0,0]

    if moveOut in Q2:
        Q2[moveOut][1] += 1
    else:
        Q2[moveOut] = [0,1,0]
```

```python
#renewal
for index,row in dup_list.iterrows():
    moveIn = datetime.strptime(row['StartDate'], '%m/%d/%Y').date()
    moveIn = moveIn.strftime("%m/%Y")
    renewal = moveIn
    if renewal in Q2:
        Q2[renewal][2] += 1
    else:
        Q2[renewal] = [0,0,1]
```

```python
#renewal-remove duplicates
drop_list=dup_list.drop_duplicates(subset='CustomerNo',keep='first', inplace=False)
for index,row in drop_list.iterrows():
    moveIn = datetime.strptime(row['StartDate'], '%m/%d/%Y').date()
    moveIn = moveIn.strftime("%m/%Y")
    renewal = moveIn
    if renewal in Q2:
        Q2[renewal][2] -= 1
```

```python
# This is the Q2 answer:
Q2 = pd.DataFrame.from_dict(Q2, orient='index')
Q2.columns=['MoveIn', 'MoveOut','Renewal']
print(Q2.head(20))
```

```python
Q2['Year'] = pd.to_datetime(Q2.index).year
#Annual Statistic:
```

```python
Q2_Annual = Q2.groupby(['Year'],as_index=False).agg({'MoveIn': 'sum', 'MoveOut':'sum','Renewal':'sum'})
Q2_Annual
```

### Using SARIMA model to build the time series model because the model is seasonal 

There are four seasonal elements that are not part of ARIMA that must be configured; they are:

P: Seasonal autoregressive order.

D: Seasonal difference order.

Q: Seasonal moving average order.

m: The number of time steps for a single seasonal period.

Forecast the monthly occupancy for Studios

```python
from math import sqrt
from sklearn.metrics import mean_squared_error

#data set
data = y1

# split dataset into train/test sets
train = data[:48]
test = data[48:]
test = [x for x in test]
predictions = []
    
history = [x for x in train]
for i in range(len(test)):
    # fit model and make forecast for history
    model = SARIMAX(data, order=(1, 0, 0), seasonal_order=(1, 1, 1, 1))
    model_fit = model.fit(disp=False)
    # make prediction
    yhat = model_fit.predict(len(history), len(history))
    # store forecast in list of predictions
    predictions.append(yhat)
    # add actual observation to history for the next loop
    history.append(test[i])
    
# estimate prediction error
predictions = [x.tolist()[0] for x in predictions]
test = [x for x in test]
print("Predictions: ",predictions)
# print(test)
error = sqrt(mean_squared_error(test, predictions))
print("error is ",error)
```
